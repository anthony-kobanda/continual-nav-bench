### **B. Data**

```python
from PIL import Image

import matplotlib.pyplot as plt
import os
```

```python
seed = 0

n_episodes = [10, 50, 125, 250, 500, 1000]
```

___

### **C. Trajectories Visualization**

```python
if os.path.exists("./plots/trajectories_N{}_S{}.png".format(n_episodes[0],seed)):

    for n_episode in n_episodes:

        print("\nN Episode : {} / Seed : {}".format(n_episode,seed))
        
        # set figure size
        plt.figure(figsize=(20, 10))

        # plot the saved image
        img = Image.open("./plots/trajectories_N{}_S{}.png".format(n_episode,seed))
        plt.imshow(img)
        plt.axis('off')
        plt.show()
```

___

### **C. Metrics**

```python
# if the metrics plot already exists, we load it and show it

if os.path.exists("./plots/metrics_S{}.png".format(seed)):

    # set figure size
    plt.figure(figsize=(20, 10))

    # plot the saved image
    img = Image.open("./plots/metrics_S{}.png".format(seed))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
